```python
from bertopic import BERTopic
topic_model_with_instructor = BERTopic()
topic_model_without_instructor = BERTopic()
```

```python
import pandas as pd
docs = pd.read_excel('Fall_2021_CSCA20_piazza.xlsx', sheet_name='Merged Table')
docs_no_instructor = docs[docs['name'] != 'Brian Harrington']['content']
docs_no_instructor = [t for t in docs_no_instructor]
topics, _ = topic_model_without_instructor.fit_transform(docs_no_instructor)
```
